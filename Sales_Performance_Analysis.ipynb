# Sales Performance Analysis Using Python

Horizon TechX Internship Project

```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```

df = pd.read_csv("HorizonTechX_Sales_Dataset.csv")

print("Dataset Loaded Successfully")
df.head()
```

```

print("Shape:", df.shape)
df.info()
```

```

df.describe()
```

```

print(df.isnull().sum())
print("Duplicates:", df.duplicated().sum())

df.drop_duplicates(inplace=True)
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Month"] = df["Order_Date"].dt.month_name()
```

```

total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()

print("Total Sales:", round(total_sales,2))
print("Total Profit:", round(total_profit,2))
```

```

top_products = df.groupby("Product")["Sales"].sum().sort_values(ascending=False)
top_products
```

```

plt.figure(figsize=(8,5))
sns.histplot(df["Sales"], bins=30)
plt.title("Sales Distribution")
plt.show()
```

```

plt.figure(figsize=(10,5))
top_products.plot(kind="bar")
plt.title("Top Selling Products")
plt.show()
```

```

category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")
plt.title("Category Wise Sales")
plt.show()
```

```

region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))
region_sales.plot(kind="bar")
plt.title("Regional Sales")
plt.show()
```

```

monthly_sales = df.groupby("Month")["Sales"].sum()

plt.figure(figsize=(12,5))
monthly_sales.plot(marker="o")
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.show()
```


## Business Insights

1. Technology products generate the highest revenue.
2. Laptops and Smartphones are top-selling products.
3. Revenue is distributed across all regions.
4. Sales and Profit show a positive correlation.
5. High-value orders contribute significantly to revenue.
6. Technology category dominates overall sales.
7. Regional demand varies across markets.
8. Monthly sales trends indicate seasonal patterns.
9. Quantity sold impacts overall profitability.
